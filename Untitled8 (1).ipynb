```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

```

```python
# Dataset directory
data_dir = "C:/Users/hp/Downloads/archive (1)/LEGO brick images v1"

# Dimensions for resizing images
img_width, img_height = 150, 150

# Batch size for training and validation
batch_size = 32

# Number of epochs for training
epochs = 10

# Path to the label file
label_file = "C:/Users/hp/Downloads/archive (1)/validation.txt"
```

```python
# Data generator for training set
train_datagen = ImageDataGenerator(
    rescale=1.0/255,  # Rescale pixel values to [0, 1]
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Splitting training data for validation
)

# Data generator for validation set
val_datagen = ImageDataGenerator(rescale=1.0/255)

# Load and preprocess training images
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'  # Specify this is for training
)

# Load and preprocess validation images
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'  # Specify this is for validation
)

```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(16, activation='softmax')  # Adjust the number of output classes accordingly
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)
```

```python
import matplotlib.pyplot as plt

# Extract test loss and accuracy from history
test_loss = history.history['val_loss']
test_accuracy = history.history['val_accuracy']

# Plot test loss
plt.figure(figsize=(10, 5))
plt.plot(test_loss, label='Test Loss', color='red')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Test Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Plot test accuracy
plt.figure(figsize=(10, 5))
plt.plot(test_accuracy, label='Test Accuracy', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Test Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# Dataset directory
data_dir = "C:/Users/hp/Downloads/archive (1)/LEGO brick images v1"

# Dimensions for resizing images
img_width, img_height = 150, 150

# Batch size for training and validation
batch_size = 32

# Number of epochs for training
epochs = 10

# Path to the label file
label_file = "C:/Users/hp/Downloads/archive (1)/validation.txt"
```

```python
# Data generator for training set
train_datagen = ImageDataGenerator(
    rescale=1.0/255,  # Rescale pixel values to [0, 1]
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Splitting training data for validation
)
# Data generator for validation set
val_datagen = ImageDataGenerator(rescale=1.0/255)

# Load and preprocess training images
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',  # Specify this is for training
    seed=42
)
```

```python
# Load and preprocess validation images
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',  # Specify this is for validation
    seed=42
)
```

```python
from tensorflow.keras.layers import Input

# Define AlexNet model
model_alexnet = Sequential([
    Conv2D(32, (11, 11), strides=(1, 1), activation='relu', input_shape=(img_width, img_height, 3)),
    MaxPooling2D((3, 3), strides=(2, 2)),
    Conv2D(64, (5, 5), activation='relu'),
    MaxPooling2D((3, 3), strides=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    Conv2D(128, (3, 3), activation='relu'),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((3, 3), strides=(2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(16, activation='softmax')  # Adjust the number of output classes accordingly
])
```
